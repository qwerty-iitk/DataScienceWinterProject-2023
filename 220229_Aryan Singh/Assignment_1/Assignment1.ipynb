### Q1. Matrix Multiplication

##### (i) Check if matrix multiplication is valid

```python
def isValid(A,B):
    if len(A[0])==len(B):
      return 1
    else:
        return 0
```

##### (ii) Using loops (without using numpy)

```python
def matrix_multiply(A,B):
    C = [[0 for _ in range(len(B[0]))]for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
          for k in range(len(A[0])):
            C[i][j] = C[i][j] + A[i][k] * B[k][j]
    return C
```

##### (iii) Using numpy

```python
def matrix_multiply_2(A,B):
    matrix_A = np.array(A)
    matrix_B = np.array(B)
    C = np.dot(matrix_A, matrix_B)
    return C
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
    total_sum = sum(x)
    mean = total_sum/len(x)
    return mean
```

```python
def standard_deviation(x):
    mean_x = mean(x)
    variance = (sum(pow(i-mean_x,2) for i in x))/len(x)
    std_dev = pow(variance, 0.5)
    return std_dev
```

```python
def zscore_normalisation(x):
    mean_x = mean(x)
    std_dev_x = standard_deviation(x)
    new_x = [((i-mean_x)/std_dev_x) for i in x]
    return new_x
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
    array_x = np.array(x)
    mean_x = np.mean(array_x)
    std_dev_x = np.std(array_x)
    new_x = (array_x-mean_x)/std_dev_x
    return new_x
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
    array_x = np.array(x)
    sigmoid = 1/(1 + np.exp(-x))
    return sigmoid
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
    sigmoid_x = sigmoidfn(x)
    derivative_sigmoid_x = sigmoid_x * (-1 + sigmoid_x)
    return derivative_sigmoid_x
```

#Linear Regression
You're given a 100 data points generated from the function
$y=mx+c$. Some noise is added to the y values.

**Task:** Estimate the values of m and c using linear regression. Write a suitable cost function and update the parameters.

```python
 # Generation of data points
x = np.random.rand(100)*100
y = -4*x+15
y = y+np.random.randn(100)*30
plt.scatter(x,y)
```

```python
# Initialisng parameters m and c such that the best fit line is given by y = m*x+c
m = 1
c = 0
#Cost function to be minimised
def compute_cost(X, y, m, c):
    m_examples = len(y)
    predictions = m * X + c
    cost = (1 / (2 * m_examples)) * np.sum((predictions - y) ** 2)
    return cost
#Function for updating parameter
def update_parameters(X, y, m, c, learning_rate):
    m_examples = len(y)
    predictions = m * X + c

    # Update parameters using gradient descent
    m_gradient = (1 / m_examples) * np.sum((predictions - y) * X)
    c_gradient = (1 / m_examples) * np.sum(predictions - y)

    m -= learning_rate * m_gradient
    c -= learning_rate * c_gradient

    return m, c
```

Run the model

```python
# define number of iterations for minimising cost function
iterations = 100
# define the learning rate
learning_rate = 0.0001
# run the model
def linear_regression(m , c, X, y, learning_rate, num_iterations):
    costs = []

    for _ in range(num_iterations):
        cost = compute_cost(X, y, m, c)
        costs.append(cost)

        m, c = update_parameters(X, y, m, c, learning_rate)

    return m, c, costs
m, c, cost = linear_regression(m, c, x, y, learning_rate, iterations)
```
