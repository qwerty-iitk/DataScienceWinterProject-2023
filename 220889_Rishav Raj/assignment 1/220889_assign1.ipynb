```python
def isValid(A,B):
  row_A,col_A=A.shape
  row_B,col_B=B.shape
  return col_A==row_B
```

```python
def matrix_multiply(A,B):
  row_A, col_A = A.shape
  row_B, col_B = B.shape

  c = np.zeros((row_A, col_B))

  for i in range(row_A):
    for j in range(col_B):
      for k in range(col_A):
        c[i][j] += A[i][k] * B[k][j]
  print(c)
```

### Q2. Z-Score Normalisation

Z-score normalization refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

New value = (x – μ) / σ

where:

x: Original value

μ: Mean of data

σ: Standard deviation of data

##### (i) Without using numpy

```python
def mean(x):
  a1=0
  for i in range(len(x)):
      a=x[i]
      a1=a1+a
  return a1/len(x)
```

```python
def standard_deviation(x):
  a1=0
  for i in range(len(x)):
    a=((x[i]-mean(x))**2)/len(x)
    a1=a1+a
  return a1 ** 0.5
```

```python
def zscore_normalisation(x):
  a=[0]*len(x)
  a1=0
  for i in range(len(x)):
    a[i]=(x[i]-mean(x))/standard_deviation(x)
  return a
```

##### (ii) Using numpy

Numpy has in_built functions for calculating mean and standard deviation

```python
def zscore_normalisation_2(x):
  a=[0]*len(x)
  mean_x=np.mean(x)
  std_dev=np.std(x)
  for i in range(len(x)):
    a[i]=(x[i]-mean_x)/std_dev
  return a
```

### Q3. Sigmoid fn and its derivative

The sigmoid function is a mathematical function that maps any input value to a value between 0 and 1.

It is defined mathematically as s(x) = 1/(1+e^(-x)).

##### (i) Write a fn to implement sigmoid fn

```python
def sigmoidfn(x):
  e=2.718
  row_x,col_x=x.shape
  s=np.zeros((row_x,col_x))
  for i in range(row_x):
    for j in range(col_x):
      s[i,j]=1/(1+e**(-x[i,j]))
  return s
```

##### (ii) Write a fn to implement derivative of sigmoid fn

```python
def derivative(x):
  e=2.718
  row_x,col_x=x.shape
  d=np.zeros((row_x,col_x))
  for i in range(row_x):
    for j in range(col_x):
      d[i,j]=-e**(-x[i,j])/((1+e**(-x[i,j]))**2)
  return d
```

#Linear Regression
You're given a 100 data points generated from the function
$y=mx+c$. Some noise is added to the y values.

**Task:** Estimate the values of m and c using linear regression. Write a suitable cost function and update the parameters.

```python
 # Generation of data points
x = np.random.rand(100)*100
y = -4*x+15
y = y+np.random.randn(100)*30
plt.scatter(x,y)
```

```python

x = x.reshape(-1, 1)
# Step 1: Initializing parameters m and c for y = m*x + c
m = 0  # Initial slope
c = 0  # Initial intercept

# Step 2: Define the cost function to be minimized (Mean Squared Error)
def cost_function(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Step 3: Function for updating parameters using gradient descent
def update_parameters(x, y_true, y_pred, m, c, learning_rate):
    dm = np.mean((y_pred - y_true) * x)  # Partial derivative of cost wrt m
    dc = np.mean(y_pred - y_true)  # Partial derivative of cost wrt c
    
    m -= learning_rate * dm  # Update m
    c -= learning_rate * dc  # Update c
    
    return m, c
```

Run the model

```python
# Step 4: Define number of iterations for minimizing the cost function
num_iterations = 100000

# Step 5: Define the learning rate
learning_rate = 0.0005  # Decrease the learning rate

# Step 6: Run the model
for i in range(num_iterations):
    y_pred = m * x + c  # Predicted values
    cost = cost_function(y, y_pred)  # Calculate the cost
    
    m, c = update_parameters(x, y, y_pred, m, c, learning_rate)  # Update parameters
    
    if i % 10000 == 0:  # Print cost every 1000 iterations
        print(f"Iteration {i}: Cost = {cost:.4f}")
```
